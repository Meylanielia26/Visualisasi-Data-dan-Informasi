```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', hue='Species', palette='Set1', s=100)
plt.title('Petal Length vs Petal Width by Iris Species', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=12)
plt.ylabel('Petal Width (cm)', fontsize=12)
plt.show()
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=df, palette='Set2')
plt.title('Distribution of Petal Length by Species', fontsize=16)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal Length (cm)', fontsize=12)
plt.show()
```

```python
sns.pairplot(df, hue="Species", palette="Set1", diag_kind="kde", markers=["o", "s", "D"])
plt.suptitle('Pair Plot of Iris Dataset', y=1.02, fontsize=16)
plt.show()
```
